# Obtendo os dados

   Utilizamos o Google Finance (https://finance.google.com/finance?q=bvmf:jbss3) para adquirir as ações da JBS e também utilizamos uma API (https://github.com/GeneralMills/pytrends) para obter o número de buscas da JBS durante o ano de 2017. 

```python
import numpy as np
import pandas as pd
import json
from pytrends.request import TrendReq
import datetime
import seaborn as sns

%matplotlib inline

pytrends = TrendReq(hl='pt-BR', tz=360)
x = pytrends.suggestions('JBS')
kw_list = [x[0]['mid']]

jbss3 = pd.read_csv('jbss3.csv')
twitter = pd.read_csv('tweets_related_per_week.csv', sep=';')
twitter_day = pd.read_csv('tweets_related_per_day.csv', sep=';')
```

```python
def getTrends(x, y):
    str = x.__str__() + ' ' + y.__str__()
    pytrends.build_payload(kw_list, cat=0, timeframe=str , geo='', gprop='')
    df = pytrends.interest_over_time()
    df.columns = [ 'value', 'isPartial'] 
    df.drop('isPartial', axis=1, inplace=True)
    return df
```

# Pré-processamento dos dados

```python
new_format = '%d-%b-%y'

f = lambda x: datetime.datetime.strptime(x, str)

jbss3['Date'] = jbss3['Date'].apply(f)
jbss3.set_index('Date',inplace=True)

jbss3 = jbss3.resample('W', closed='left', label='left').mean()

trends = getTrends(datetime.date(2017,1,1), datetime.date.today())

jbss3 = jbss3.join(trends)

jbss3['Previous value'] = jbss3['value'].shift()
```

```python
twitter['date'] = pd.to_datetime(twitter['date'], format='%d/%m/%Y')
twitter.set_index('date', inplace=True)

jbss3 = jbss3.join(twitter)

jbss3['Previous Tweets'] = jbss3['tweets'].shift()
```

```python

jbss3['Relative Change value'] = (jbss3['Previous value'] - jbss3['Previous value'].shift())/jbss3['Previous value'].shift()
jbss3['Relative Change tweets'] = (jbss3['Previous Tweets'] - jbss3['Previous Tweets'].shift())/jbss3['Previous Tweets'].shift()
jbss3['Relative Change Close'] = (jbss3['Close'] -jbss3['Close'].shift())/ jbss3['Close'].shift()

jbss3.head()
```

# Análise dos dados

Vamos nos restringir a uma janela (26/02/2017 a 02/10/2017) menor do que um ano e que cobre os três grandes picos.

```python
import matplotlib.pyplot as plt

def plotClose(str):
    jbss = jbss3.copy()
    jbss.reset_index(inplace=True)
    x = jbss['Date']
    y1 = jbss[str]
    y2 = jbss['Close']

    fig, ax1 = plt.subplots()

    ax2 = ax1.twinx()
    ax1.plot(x, y1, 'g-')
    ax2.plot(x, y2, 'b-')

    ax1.set_xlabel('Tempo')
    ax1.set_ylabel(str, color='g')
    ax2.set_ylabel('Close', color='b')
```

```python
start = datetime.date(2017, 2, 26)
end = datetime.date(2017, 9,23)
jbss3 = jbss3[start:end]


fig, ax = plt.subplots()
ax.plot(jbss3['Close'], 'g-')
ax2 =ax.twinx()
ax2.plot(jbss3['value'])
sns.despine(ax=ax, right=True, left=True)
sns.despine(ax=ax2, left=True, right=False)
ax2.spines['right'].set_color('white')


fig, ax = plt.subplots()
ax.plot(jbss3['Close'], 'g-')
ax2 =ax.twinx()
ax2.plot(jbss3['tweets'])
sns.despine(ax=ax, right=True, left=True)
sns.despine(ax=ax2, left=True, right=False)
ax2.spines['right'].set_color('white')
```

# Estimando os Dados com Regressão Linear

```python
def linearRegressionPredict(X,y):
    X_train, X_test, y_train, y_test = train_test_split(
          X, y, random_state=0, test_size=.33)

    lr = LinearRegression()

    model = lr.fit(X_train[:,np.newaxis], y_train)

    predictions = model.predict(X_test[:,np.newaxis])
    print('Rˆ2 is:', r2_score(y_test , predictions))
    print('RMES is:', mean_squared_error(y_test, predictions))
```

# Estimando os dados com Random Forest

```python
def randomForestPredict(X,y):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, random_state=0, test_size=.33)


    rf = RandomForestRegressor(n_estimators=100, random_state=0)


    rf.fit(X_train[:,np.newaxis], y_train)

    prediction = rf.predict(X_test[:,np.newaxis])
    print('Rˆ2 is:', r2_score(y_test , prediction))
    print('RMES is:', mean_squared_error(y_test, prediction))
```

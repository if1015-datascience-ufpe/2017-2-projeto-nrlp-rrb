# Obtendo os dados

   Utilizamos o Google Finance (https://finance.google.com/finance?q=bvmf:jbss3) para adquirir as ações da JBS e também utilizamos uma API (https://github.com/GeneralMills/pytrends) para obter o número de buscas da JBS durante o ano de 2017. 

```python
import numpy as np
import pandas as pd
import json
from pytrends.request import TrendReq
import datetime
import seaborn as sns

%matplotlib inline

pytrends = TrendReq(hl='pt-BR', tz=360)
x = pytrends.suggestions('JBS')
kw_list = [x[0]['mid']]

jbss3 = pd.read_csv('jbss3.csv')
twitter = pd.read_csv('tweets_related_per_week.csv', sep=';')
twitter_day = pd.read_csv('tweets_related_per_day.csv', sep=';')
```

```python
def getTrends(x, y):
    str = x.__str__() + ' ' + y.__str__()
    pytrends.build_payload(kw_list, cat=0, timeframe=str , geo='', gprop='')
    df = pytrends.interest_over_time()
    df.columns = [ 'searchs', 'isPartial'] 
    df.drop('isPartial', axis=1, inplace=True)
    return df
```

# Pré-processamento dos dados

No pré-processamento dos dados, foi escolhido dividir o conjunto em semanas para tentar facilitar a análise. Também criamos outras colunas, Previous value e Previous tweets, que são os valores de pesquisar e de tweets feitos na semana anterior e que serão utilizado para prever o valor da ação atual.

```python
new_format = '%d-%b-%y'

f = lambda x: datetime.datetime.strptime(x, new_format)

jbss3['Date'] = jbss3['Date'].apply(f)
jbss3.set_index('Date',inplace=True)

jbss3 = jbss3.resample('W', closed='left', label='left').mean()

trends = getTrends(datetime.date(2017,1,1), datetime.date.today())

jbss3 = jbss3.join(trends, how='inner')
```

```python
twitter['date'] = pd.to_datetime(twitter['date'], format='%d/%m/%Y')
twitter.set_index('date', inplace=True)

jbss3 = jbss3.join(twitter, how='inner')


jbss3['Previous searchs'] = jbss3['searchs'].shift()
jbss3['Previous Tweets'] = jbss3['tweets'].shift()
```

## Criando a coluna da mudança relativa

```python

jbss3['Relative Change searchs'] = (jbss3['Previous searchs'] - jbss3['Previous searchs'].shift())/jbss3['Previous searchs'].shift()
jbss3['Relative Change tweets'] = (jbss3['Previous Tweets'] - jbss3['Previous Tweets'].shift())/jbss3['Previous Tweets'].shift()
jbss3['Relative Change Close'] = (jbss3['Close'] -jbss3['Close'].shift())/ jbss3['Close'].shift()

jbss3.head()
```

# Análise dos dados

Vamos nos restringir a uma janela (26/02/2017 a 02/10/2017) menor do que um ano e que cobre os três grandes picos.

```python
import matplotlib.pyplot as plt

def plotTimeSeries(str):
    fig, ax = plt.subplots()
    ax.plot(jbss3['Close'], 'g-')
    ax.set_ylabel('Close', color='g')
    ax2 = ax.twinx()
    ax2.plot(jbss3[str])
    ax2.set_ylabel(str)
    sns.despine(ax=ax, right=True, left=True)
    sns.despine(ax=ax2, left=True, right=False)
    ax2.spines['right'].set_color('white')
```

Fazendo uma comparação entre o preço da ação e google trend, e entre o preço da ação e os tweets. Com os gráficos abaixo, podemos perceber que para o caso da empresa JBS o preço da ação e as variáveis independentes tem o sentido contrário. 

```python
start = datetime.date(2017, 2, 26)
end = datetime.date(2017, 9,23)
jbss3 = jbss3[start:end]


plotTimeSeries('searchs')

plotTimeSeries('tweets')
```

# Estimando os Dados com Regressão Linear

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, explained_variance_score, mean_squared_error

def linearRegressionPredict(X,y):
    X_train, X_test, y_train, y_test = train_test_split(
          X, y, random_state=0, test_size=.33)

    lr = LinearRegression()

    model = lr.fit(X_train[:,np.newaxis], y_train)

    predictions = model.predict(X_test[:,np.newaxis])
    print('Rˆ2 is:', r2_score(y_test , predictions))
    print('RMES is:', mean_squared_error(y_test, predictions))
```

## Google Trend

A regressão linear não teve um bom resultado com a váriavel independete Log searchs. Uma possível causa é a presença de outliers.

```python

X = jbss3['Log searchs'].values.copy() 
y = jbss3['Log Close'].values

linearRegressionPredict(X,y)
```

## Twitter

Aqui a regressão linear teve um bom desempenho, com um alto coeficiente de determinação.

```python
X = jbss3['Log Tweets'].values.copy()


linearRegressionPredict(X,y)
```

# Estimando os dados com Random Forest

O Random Forest teve um resultado pior que a regressão linear. Talvez esse algoritmo não seja ideal para um pequena base de dados.

```python
from sklearn.ensemble import RandomForestRegressor

def randomForestPredict(X,y):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, random_state=0, test_size=.33)

    rf = RandomForestRegressor(n_estimators=100, random_state=0, criterion='mse')


    rf.fit(X_train[:,np.newaxis], y_train)

    prediction = rf.predict(X_test[:,np.newaxis])
    print('Rˆ2 is:', r2_score(y_test , prediction))
    print('RMES is:', mean_squared_error(y_test, prediction))
```

## Google Trend

```python
X = jbss3['Log searchs'].values.copy()
y = jbss3['Log Close'].values

randomForestPredict(X,y)
```

## Twitter

```python
X = jbss3['Log Tweets'].values.copy() 


randomForestPredict(X,y)
```

```python
jbss3 = jbss3.dropna()

X = jbss3['Relative Change searchs'].values.copy() 
y = jbss3['Relative Change Close'].values

linearRegressionPredict(X,y)
```
